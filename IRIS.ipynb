```python
se = df[df['Species'] == 'Iris-setosa']
ve = df[df['Species'] == 'Iris-versicolor']
vi = df[df['Species'] == 'Iris-virginica']

f, axs = plt.subplots(3,2, figsize = (15, 8))

sns.kdeplot(data = df, x = se['SepalLengthCm'], y = se['SepalWidthCm'], cmap ='plasma', shade = 'True', ax = axs[0,0])
axs[0,0].set_title("Setosa-sepal")
sns.kdeplot(data = df, x = se['PetalLengthCm'], y = se['PetalWidthCm'], cmap ='plasma', shade = 'True', ax = axs[0,1])
axs[0,1].set_title("setosa-petal")
sns.kdeplot(data = df, x = ve['SepalLengthCm'], y = ve['SepalWidthCm'], cmap ='plasma', shade = 'True', ax = axs[1,0])
axs[1,0].set_title("Versicolor-sepal")
sns.kdeplot(data = df, x = ve['PetalLengthCm'], y = ve['PetalWidthCm'], cmap ='plasma', shade = 'True', ax = axs[1,1])
axs[1,1].set_title("Versicolor-petal")
sns.kdeplot(data = df, x = vi['SepalLengthCm'], y = vi['SepalWidthCm'], cmap ='plasma', shade = 'True', ax = axs[2,0])
axs[2,0].set_title("virginica-sepal")
sns.kdeplot(data = df, x = vi['PetalLengthCm'], y = vi['PetalWidthCm'], cmap ='plasma', shade = 'True', ax = axs[2,1])
axs[2,1].set_title("Viginica-petal")

plt.tight_layout()
```
